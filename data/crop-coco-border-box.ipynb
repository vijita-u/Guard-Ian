```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import cv2
import ast
import matplotlib.pyplot as plt
from IPython.display import FileLink


# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os

csv_path = "/kaggle/input/datasets/ginoleone117/coco-data/filtered_coco_metadata.csv"
img_directory_train = "/kaggle/input/datasets/awsaf49/coco-2017-dataset/coco2017/train2017"

df = pd.read_csv(csv_path)
os.makedirs('/kaggle/working/cropped_train_images', exist_ok=True)


# You can write up to 20GB to the current directory (/kaggle/working/) that gets preserved as output when you create a version using "Save & Run All" 
# You can also write temporary files to /kaggle/temp/, but they won't be saved outside of the current session
```

```python
counter = 0

def clean_and_eval(val):
    try:
        # Remove any leading/trailing spaces or hidden characters
        cleaned = str(val).strip()
        return ast.literal_eval(cleaned)
    except:
        return None

for index, row in df.iterrows():

    #for new unique filenames
    
    #convert bbox string into a list type
    df['bbox'] = df['bbox'].apply(clean_and_eval)
    #print(row["bbox"], row["file_name"])
    img_path = os.path.join(img_directory_train, row["file_name"])
    bounding_box_image = cv2.imread(img_path)

    #convert BGR(open cv default) to RGB(matplotlib default)
    #plt.imshow(cv2.cvtColor(bounding_box_image, cv2.COLOR_BGR2RGB))
    #plt.show()

    #get all bbox metrics
    x_min, y_min, width, height = (row['bbox'][0], row['bbox'][1], row['bbox'][2], row['bbox'][3])

    start_y = int(y_min)
    end_y = int(y_min + height)
    start_x = int(x_min)
    end_x = int(x_min + width)

    cropped_img = bounding_box_image[start_y: end_y, start_x: end_x]
    #plt.imshow(cv2.cvtColor(cropped_img, cv2.COLOR_BGR2RGB))
    
    cropped_img_name = f"cropped_{counter:05d}.jpg"
    cropped_folder = "/kaggle/working/cropped_train_images"
    full_path = cropped_folder + "/" + cropped_img_name
    cv2.imwrite(full_path, cropped_img)

    #change image file path to new path in csv
    df.at[index, "file_name"] = cropped_img_name
    counter += 1
```

```python
df.to_csv("cropped.csv", index=False)
FileLink(r'cropped.csv')
```
